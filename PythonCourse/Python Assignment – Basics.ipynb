#### Task 3: Number Operations

**Write a program that:**

- Takes two numbers from the user

- Prints the result of:

Addition ,
Subtraction ,
Multiplication ,
Division ,
Modulus

```python
print("Task 3: Number Operations")
print("-" * 30)

num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))

addition = num1 + num2
subtraction = num1 - num2
multiplication = num1 * num2

if num2 != 0:
    division = num1 / num2
    modulus = num1 % num2
else:
    division = "Undefined (division by zero)"
    modulus = "Undefined (modulus by zero)"

print(f"Addition: {num1} + {num2} = {addition}")
print(f"Subtraction: {num1} - {num2} = {subtraction}")
print(f"Multiplication: {num1} Ã— {num2} = {multiplication}")
print(f"Division: {num1} Ã· {num2} = {division}")
print(f"Modulus: {num1} % {num2} = {modulus}")
```

#### Task 4: Booleans and Comparisons

**Write a program that:**

- Takes two numbers from the user

- Checks and prints whether:
The first number is greater than the second , The two numbers are equal

```python
print("Task 4: Booleans and Comparisons")
print("-" * 30)

num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))

is_greater = num1 > num2
are_equal = num1 == num2

print(f"Is {num1} greater than {num2}? {is_greater}")
print(f"Are {num1} and {num2} equal? {are_equal}")
```

#### Task 5: String Operations

**Write a program that:**

- Takes a sentence from the user

- Prints: The length of the sentence , The sentence in uppercase , The sentence in lowercase

```python
print("Task 5: String Operations")
print("-" * 30)

sentence = input("Enter a sentence: ")

length = len(sentence)
uppercase = sentence.upper()
lowercase = sentence.lower()

print(f"Length of the sentence: {length} characters")
print(f"Sentence in uppercase: {uppercase}")
print(f"Sentence in lowercase: {lowercase}")
```

#### Task 6: String Functions

**Write a program that:**

- Takes a word from the user

- Replaces all occurrences of the letter a with @

- Prints the modified word

```python
print("Task 6: String Functions")
print("-" * 30)

word = input("Enter a word: ")

modified_word = word.replace('a', '@')

print(f"Original word: {word}")
print(f"Modified word: {modified_word}")
```
